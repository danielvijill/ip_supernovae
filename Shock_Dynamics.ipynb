## Shock Dynamics

The inner density profile of the CSM closely follows $ρ ∝ r^{−1.5}$. Initially the shocked region is well approximated to be adiabatic with an index $γ = 4/3$, and a selfsimilar solution for the shock propagation in a CSM with $ρ = qr^{−1.5}$ (Chevalier 1982) gives the radius and the velocity of the shock front as functions of time as

$r_{sh} = \left( \frac{Ag^n}{q} \right)^{1/(n-1.5)} t^{(n-3)/(n-5)}$

$\nu_{sh} = \frac{dr_{sh}}{dt} = \frac{n-3}{n-1.5} \frac{r_sh}{t}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 100)
```

```python
n=8;
A=1;
g=1.8;
```

```python
Mcsm = 1 #mass of the csm in solar masses
r_out = 10e17 #radius of CSM distribution in cm
```

```python
q = 7.5*10e8 * (Mcsm/10**(-1)) * (r_out/10e15)**(-1.5)
```

```python
rsh = ((A*g**n)/q)**(1/(n-1.5)) * t**((n-3)/(n-1.5))
```

```python
plt.figure(figsize=(10,5))
plt.plot(t, rsh, color='r')
plt.grid()
plt.xlabel('$t$')
plt.ylabel('$r_{sh}$')
plt.title('Shock Radius as a function of time')
```

```python
g = 2.8
```

```python
g_values = [1, 1.8, 3, 10]
```

```python
fig = plt.figure(1)

for i in g_values:
    rsh = ((A*i**n)/q)**(1/(n-1.5)) * t**((n-3)/(n-1.5))
    plt.plot(t, rsh, label=f'g={i}')
    plt.grid()
    plt.xlabel('$t$')
    plt.ylabel('$r_{sh}$')
    plt.title('Shock Radius as a function of time and g')
    plt.legend()
```

```python
vsh = (n-3)/(n-1.5) * (rsh/t)
```

```python
plt.figure(figsize=(10,5))
plt.plot(t, vsh, color='b')
plt.grid()
plt.xlabel('$t$')
plt.ylabel('$v_{sh}$')
plt.title('Shock Velocity as a function of time')
```
